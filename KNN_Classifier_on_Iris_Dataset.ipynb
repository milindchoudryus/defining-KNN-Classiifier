```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X[:60], y[:60], test_size=30, stratify=y[:60])
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 20, stratify = y_train)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)


errors = accuracy_score(y_test, y_pred)
errors = 50 - errors

errors
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)


errors = accuracy_score(y_test, y_pred)
errors = 50 - errors

errors
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)


errors = accuracy_score(y_test, y_pred)
errors = 50 - errors

errors
```

```python
knn = KNeighborsClassifier(n_neighbors=3, metric = 'manhattan')
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)


errors = accuracy_score(y_test, y_pred)
errors = 50 - errors

errors
```

```python
knn = KNeighborsClassifier(n_neighbors=5, weights = 'distance')
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)


errors = accuracy_score(y_test, y_pred)
errors = 50 - errors

errors
```
